## Exploratory Data Analysis

```python
ax=sns.countplot(x='Gender',data=data)

for label in ax.containers:
  ax.bar_label(label)
```

```python
sales_data=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_data)
```

From above graph we can conclude that number of buyers in diwali are mostly females

Now we have to find the customers age

```python
data.columns
```

```python
ax=sns.countplot(data=data,x='Age Group',hue='Gender')
for label in ax.containers:
  ax.bar_label(label)
```

From this we can conclude than 26-35 age group female are the mostly buyers

```python
# Total amount vs age group
sales_age=data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
# ax=sns.barplot(x='Age Group',y='Amount',data=sales_age)
# for label in ax.containers:
#      ax.bar_label(label)
```

Now we find in which state the customers are more

```python
data.columns
```

```python
# Total orders from top 10 states
sales_state=data.groupby('State',as_index=False)['Orders'].sum().sort_values(by='State',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
# total amount per state
amount_state=data.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=amount_state)
```

Most of the orders from U.P

```python
ax=sns.countplot(data=data,x='Shaadi')
sns.set(rc={'figure.figsize':(6,5)})
for label in ax.containers:
  ax.bar_label(label)
```

```python
# sales by martial status ,gender vs amount
sales_state=data.groupby(['Shaadi','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Shaadi',y='Amount',hue='Gender')
```

Married women has more purchasing power

Occupation

```python
sales_occup=data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_occup,x='Occupation',y='Amount')
```

Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=data,x='Product_Category')

for label in ax.containers:
  ax.bar_label(label)
```

```python
sales_pdt=data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.barplot(data=sales_pdt,x='Product_Category',y='Amount')
sns.set(rc={'figure.figsize':(50,40)})
```

```python
sns.set(rc={'figure.figsize':(20,5)})
data.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
